# time series plot

```python
import matplotlib.pyplot as plt
from datetime import datetime, timedelta
import matplotlib.dates as mpl_dates

plt.style.use('seaborn')
```

```python
dates = [
    datetime (2019, 5, 25),
    datetime (2019, 5, 26),
    datetime (2019, 5, 27),
    datetime (2019, 5, 28),
    datetime (2019, 5, 29),
    datetime (2019, 5, 30)
]

y = [1, 3, 4, 6, 5, 7]
```

```python
plt.plot(dates, y)

plt.tight_layout()
plt.show()
```

```python
plt.plot(dates, y)
plt.gcf().autofmt_xdate()

plt.tight_layout()
plt.show()
```

```python
plt.plot(dates, y)
plt.gcf().autofmt_xdate()
date_format = mpl_dates.DateFormatter('%b, %d   %y')
plt.gca().xaxis.set_major_formatter(date_format)

plt.tight_layout()
plt.show()
```
